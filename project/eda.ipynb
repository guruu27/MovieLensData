```python
import requests
from bs4 import BeautifulSoup

def get_imdb_id(movie_title):
    query = f"{movie_title} site:imdb.com"
    url = f"https://www.google.com/search?q={query.replace(' ', '+')}"
    
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }
    
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find the first link that matches IMDb's title format
    for link in soup.find_all('a', href=True):
        href = link['href']
        if '/title/' in href:
            start_idx = href.find('/title/')
            end_idx = href.find('/', start_idx + 7)
            return href[start_idx + 7:end_idx]
    
    return None

# Example usage:

for movie in movie_titles:
    imdb_id = get_imdb_id(movie)
    if imdb_id:
        print(f"{movie}: https://www.imdb.com/title/{imdb_id}/reviews/")
```
